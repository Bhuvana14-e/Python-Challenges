

 Dictionary, Set, Frozenset Project Title: **Student Grades & Hobbies Tracker Scenario:** A student activity coordinator is collecting information about student grades and hobbies. They want to ensure grades are updated properly and hobby lists are stored efficiently. Tasks: Create a dictionary with student names as keys and their grades as values.

Allow the user to add or update grades for at least 3 students.

Create a set of hobbies and allow the user to input at least 3 hobbies.

Convert the set of hobbies into a frozenset and display it.

```python
student_grades = {
    "Sanjai": "A",
    "Nithya": "B+",
    "Kavi": "A-"
}

print(student_grades)
```

```python
student_grades = {}

# Add or update 3 students
name1 = input("Enter student 1 name: ")
grade1 = input("Enter grade: ")
student_grades[name1] = grade1

name2 = input("Enter student 2 name: ")
grade2 = input("Enter grade: ")
student_grades[name2] = grade2

name3 = input("Enter student 3 name: ")
grade3 = input("Enter grade: ")
student_grades[name3] = grade3

print(student_grades)
```

```python
hobbies = set()

h1 = input("Enter hobby 1: ")
hobbies.add(h1)

h2 = input("Enter hobby 2: ")
hobbies.add(h2)

h3 = input("Enter hobby 3: ")
hobbies.add(h3)

print(hobbies)

```

```python
hobbies = {h1, h2, h3}
print(hobbies)

frozen_hobbies = frozenset(hobbies)
print(frozen_hobbies)
```
