**Coding challenge 4** List Methods and Slicing Project Title: Contact Directory

Scenario: You have been assigned to build a contact management system where employees can store and retrieve contact details efficiently.

Tasks: Create a 2D list where each row holds a contact’s name, phone number, and email address. Allow the user to add a new contact. Display all contacts sorted alphabetically by name. Allow the user to search for contacts by name. The search should support partial matches. After sorting, display only the first 3 contacts in the list.

```python
# Contact Directory - 2D List
contacts = [
    ["Sanjai", "9344496347", "sanjai@gmail.com"],
    ["Nithya", "9876340988", "Nithya@gmail.com"],
    ["Karthik", "9887655500", "karthik@gmail.com"]
]

# 1️⃣ Add a new contact (input version can be added later)
new_contact = ["Kavi", "8005521675", "Kavi@gmail.com"]
contacts.append(new_contact)

print("\nAfter adding new contact:")
for c in contacts:
    print(c)

# 2️⃣ Sort contacts alphabetically by name
contacts.sort(key=lambda x: x[0])

print("\nSorted Contacts:")
for c in contacts:
    print(c)

# 3️⃣ Search by name (partial match)
search_name = "Sa"   # Example partial search
print(f"\nSearch results for '{search_name}':")

for c in contacts:
    if search_name.lower() in c[0].lower():
        print(c)

# 4️⃣ Display only first 3 contacts (slicing)
print("\nFirst 3 contacts:")
for c in contacts[:3]:
    print(c)
```
