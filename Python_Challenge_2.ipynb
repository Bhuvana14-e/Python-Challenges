```python
secret_msg=input("Enter your secret message: ")
uppercase=secret_msg.upper()
lowercase=secret_msg.lower()
print("secret message converted to uppercase", uppercase,"\n secret message converted to lowercase", lowercase)
print("Replaced all spaces with underscore-",secret_msg.replace(" ",""))
print("Every 3rd charcter of the secret msg",secret_msg[::3],)
print("Reversed the message",secret_msg[::-1])
print("Length of the secret message:",len(secret_msg))
print("Secret message is alphabet-",secret_msg.isalpha())
```
